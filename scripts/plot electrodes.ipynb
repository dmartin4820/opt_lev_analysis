```python
df = bu.DataFile()
```

```python
import matplotlib.mlab
```

```python
fnames = ["/data/20180802/elec_test/elec0_elec0_10000mV41Hz0mVdc.h5", "/data/20180802/elec_test/elec1_elec1_10000mV41Hz0mVdc.h5",\
          "/data/20180802/elec_test/elec2_elec2_10000mV41Hz0mVdc.h5", "/data/20180802/elec_test/elec3_elec3_10000mV41Hz0mVdc.h5",\
          "/data/20180802/elec_test/elec4_elec4_10000mV41Hz0mVdc.h5", "/data/20180802/elec_test/elec5_elec5_10000mV41Hz0mVdc.h5",\
          "/data/20180802/elec_test/elec6_elec6_10000mV41Hz0mVdc.h5", "/data/20180802/elec_test/elec7_elec7_10000mV41Hz0mVdc.h5"]
```

```python
for i, f in enumerate(fnames):
    df.load(f)
    psd, freqs = matplotlib.mlab.psd(df.electrode_data[i], Fs = 5000, NFFT = 2**14)
    plt.loglog(freqs, psd, label = "electrode " + str(i))
plt.legend()
plt.xlabel("frequency [Hz]")
plt.ylabel(r"PSD[$V^{2}/Hz$]")
plt.show()
```
